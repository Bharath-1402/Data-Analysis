**1.How many people of each race are represented in this dataset? This should be a Pandas series with race names as the index labels. (race column)**

```python
race_count = pd.Series(df["race"]).value_counts()
race_count
```

**2.What is the average age of men?**

```python
average_age_men = round(df[df["sex"] == "Male"]["age"].mean(),1)
average_age_men
```

**4.What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?**

```python
# Calculate the number of people with each higher education level
bachelors = df[df['education'] == 'Bachelors']
masters = df[df['education'] == 'Masters']
doctorates = df[df['education'] == 'Doctorate']

# Combine the dataframes for all higher education levels
higher_edu = pd.concat([bachelors, masters, doctorates])

# Calculate the percentage of people with higher education who earn >50K
higher_education_rich1 = round(((higher_edu['salary'] == '>50K').sum() / len(higher_edu)) * 100, 1)
higher_education_rich1
```

```python
#simplified
higher_education = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
higher_education_rich = round(((higher_education1['salary'] == '>50K').sum() / len(higher_education1)) * 100, 1)
higher_education_rich
```

**5.What percentage of people without advanced education make more than 50K?**

```python
lower_education = df[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
lower_education_rich = round(((lower_education['salary'] == '>50K').sum() / len(lower_education)) * 100, 1)
lower_education_rich
```

**6.What is the minimum number of hours a person works per week?**

```python
hours = pd.Series(df['hours-per-week'])
min_work_hours = hours.min()
min_work_hours
```

**7.What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?**

```python
min_hours = df[df['hours-per-week'] == min_work_hours]
rich_percentage = round(((min_hours['salary'] == '>50K').sum() / len(min_hours)) * 100, 1)
rich_percentage
```

**8.What country has the highest percentage of people that earn >50K and what is that percentage?**

```python
highest_earning_country = (df.loc[df['salary'] == '>50K', 'native-country'].value_counts() / df['native-country'].value_counts()).idxmax()
highest_earning_country_percentage = round((df.loc[(df['native-country'] == highest_earning_country) & (
df['salary'] == '>50K')].shape[0] / df.loc[df['native-country'] == highest_earning_country].shape[0]) * 100, 1)
print(highest_earning_country)
print(highest_earning_country_percentage)
```

The idxmax( ) function is used to find the
index label corresponding to the maximum value in a Pandas Series or DataFrame.

For example, in a Series, it returns the index of the element with the highest value. In a DataFrame, when applied to a column, it returns the index of the row containing the maximum value for that column.

Shape[ ] is used to get number of rows/column in the dataframe. 0 - indicies no.of rows, 1 - indicies no.of column

```python
#eg: idxmax()
c = df['native-country'].value_counts()
#print(c)
c = c.idxmax()
print(c)
```

**9.Identify the most popular occupation for those who earn >50K in India.**

```python
india = df[df['native-country'] == 'India']
india_rich = india[india['salary'] == '>50K']
top_IN_occupation = india_rich['occupation'].value_counts().idxmax()
top_IN_occupation
```

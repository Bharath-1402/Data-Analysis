```python
import numpy as np
def calculate(list):
  if len(list) != 9:
    raise ValueError("List must contain nine numbers.")
  ls = np.array(list).reshape(3,3)
  cal = {}
  cal['mean'] = [ls.mean(axis=0).tolist(),ls.mean(axis=1).tolist(),ls.mean().tolist()]
  cal['variance'] = [ls.var(axis=0).tolist(),ls.var(axis=1).tolist(),ls.var().tolist()]
  cal['standard deviation'] = [ls.std(axis=0).tolist(),ls.std(axis=1).tolist(),ls.std().tolist()]
  cal['max'] = [ls.max(axis=0).tolist(),ls.max(axis=1).tolist(),ls.max().tolist()]
  cal['min'] = [ls.min(axis=0).tolist(),ls.min(axis=1).tolist(),ls.min().tolist()]
  cal['sum'] = [ls.sum(axis=0).tolist(),ls.sum(axis=1).tolist(),ls.sum().tolist()]
  return cal
```

tolist() is used convert the elements into the list

```python
print(calculate([0,1,2,3,4,5,6,7,8]))
```

So in this project, we get the 9 elements from the user and coverted it into the 3x3 matrix. Now we need to find the values of mean, var, std, max, min, sum of this matrix by each axis(1,0) and the overall values too. And these values are stored in the dictionary
